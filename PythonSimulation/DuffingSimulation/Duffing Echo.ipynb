```python
from lib.DuffingEnsemble import DuffingEnsemble
import numpy as np
from astropy import units as u
import matplotlib.pyplot as plt
```

```python
enbl = DuffingEnsemble(20000, 1.0*u.s**-2, 0.01*u.m**-2*u.s**-2, 0.00001*u.m**2/u.s**2)
avg_pos = np.zeros(3500)
```

```python
avg_pos[0:500] = enbl.get_average_position(50.0*u.s, 0.1*u.s)
plt.plot(avg_pos)
```

```python
enbl.propagate(50.0*u.s)
enbl.dipole_kick(1.0*u.m)
```

```python
avg_pos[500:1500] = enbl.get_average_position(100.0*u.s, 0.1*u.s)
plt.plot(avg_pos)
```

```python
enbl.propagate(100.0*u.s)
enbl.quadrupole_kick(0.1/u.s)
```

```python
avg_pos[1500:3500] = enbl.get_average_position(200.0*u.s, 0.1*u.s)
plt.plot(avg_pos)
```

```python
fig, ax = plt.subplots()
ax.plot(np.linspace(0.0, 350.0, 3500)*u.s, avg_pos*u.m)
ax.set_title("Duffing Echo")
ax.set_ylabel("Average Position (m)")
ax.set_xlabel("Time (s)")

fig.savefig("DuffingEcho.pdf")
fig.show()
```
